### Categorical data

```python
# payment accepted in different branches

plt.figure(figsize=(10,4))
sns.countplot('Payment', data=data_sm, hue='Branch')
plt.title("Payments used in different branches", fontsize=15)
plt.xlabel('Product line', size=15)
plt.ylabel('Count', size= 15)
plt.xticks(rotation=90)
plt.show()
```

```python
#branch vs customer vs sales

b_c_s= pd.pivot_table(data= data_sm, columns='Branch',index='Customer type', values='Total', aggfunc=np.sum )
```

```python
sns.heatmap(b_c_s, annot=True, cmap='RdYlGn', fmt='f')
plt.show()
```

```python
#branch vs product vs sales

b_p_s= pd.pivot_table(data= data_sm, columns='Branch',index='Product line', values='Total', aggfunc=np.sum )
```

```python
sns.heatmap(b_p_s, annot=True, cmap='RdYlGn', fmt='f')
plt.show()
```

```python
#branch vs product vs rating

b_p_r= pd.pivot_table(data= data_sm, columns='Branch',index='Product line', values='Rating', aggfunc=np.mean )
```

```python
sns.heatmap(b_p_r, annot=True, cmap='RdYlGn', fmt='f')

plt.show()
```

```python
#branch vs product vs hours

b_p_h= pd.pivot_table(data= data_sm, columns='Product line',index='Customer type', values='Hours', aggfunc= lambda x:x.count())
```

```python
sns.heatmap(b_p_h, annot=True, cmap='RdYlGn', fmt='f')

plt.show()
```
