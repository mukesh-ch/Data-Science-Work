```python
# import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Remove the rows: (% Growth over the previous year)' and 'GSDP - CURRENT PRICES (in Crore) for the year 2016-17.

gdp=gdp[-(gdp['Duration'] =='2016-17')]
```

```python
gdp.head()
```

```python
# Check the total number of null values in each columns
gdp.isnull().sum()
```

```python
# Check if any column has all the values as NAN
gdp.isnull().all(axis=0)
```

```python
# removing West Bengal as the whole column is NAN
gdp=gdp.drop('West Bengal1', axis=1)
```

```python
gdp.columns
```

```python
#Calculating the average growth of states for the duration 2013-14, 2014-15 and 2015-16 by taking the mean of the row ‘(% Growth over previous year)’.
gdp
```

```python
avg_growth = gdp.iloc[6:]
```

```python
#mean of all the states
avg_growth_values = avg_growth.iloc[:,2:34].mean()
```

```python
#sort values

avg_growth_values = avg_growth_values.sort_values()
```

```python
type(avg_growth_values)
```

```python
plt.figure(figsize=(12,10), dpi = 300)
avg_growth_values.plot.barh()
plt.xlabel('Average Growth Rate', fontsize=12)
plt.ylabel('States', fontsize=12)
plt.title('Average Growth Rate for all the states',fontsize=13)
plt.show()
```

```python
average_growth_rate = avg_growth_values.to_frame(name='Average growth rate')
```

```python
average_growth_rate
```

```python
plt.figure(figsize=(12,10), dpi = 300)

sns.barplot(x = average_growth_rate['Average growth rate'], y = average_growth_rate.index,palette='viridis')
plt.xlabel('Average Growth Rate', fontsize=12)
plt.ylabel('States', fontsize=12)
plt.title('Average Growth Rate for all the states',fontsize=13)
plt.show()
```

#### Observations:

- We can see an interesting observation from the above plot, the average growth rate has been the maximum for the North East states except for Assam and Meghalaya which is not what we generally expect so we should take a further look at these states.
- The average growth rate has been least for states like Goa, Odisha, Meghalaya, Sikkim, Jammu & Kashmir etc.`

```python
# top 5 states as per average growth rate
average_growth_rate.tail()
```

```python
avg_growth[['Mizoram','Tripura','Nagaland','Manipur','Arunachal Pradesh']]
```

- We can see that the growth rate for the above states actually decreased substantially for the year 14-15 in comparison to the year 13-14 but as the growth rate was very high for the year 13-14 so the average is higher for these states.
- In the absence of data for the year 2015-16 we cannot say definitively that these are high performing states as their growth rate decreased for the year 2014-15.

To find out the states that have been growing continuously fast we need to take a look at the Standard Deviation and the Mean growth rate for the states.

```python
#create a dataframe to store the mean and the standard deviation of the growth rate for various states

describe = pd.DataFrame(avg_growth.describe())
```

```python
describe=describe.T
```

```python
describe
```

```python
describe[(describe['mean']>12) & (describe['std']<2)]
```

```python
describe[(describe['mean']>12) & (describe['std']<2)].index[0:5].to_list()
```

```python
# states having mean growth rate less than 12 and standard deviation greater than 2
describe[(describe['mean']<12) & (describe['std']>2)].sort_values(by='mean',ascending=True)
```

By comparing the average growth rate for the year 2013-14, 2014-15, 2015-16 and the standard deviation.

States that are growing consistently fast are:

```python
describe[(describe['mean']>12) & (describe['std']<2)].index[0:5].to_list()
```

States that are growing consistently struggle are:

```python
describe[(describe['mean']<12) & (describe['std']>2)].sort_values(by='mean',ascending=True).index[0:5].to_list()
```

### Plotting the total GDP of the states for the year 2015-16

```python
gdp_2015_16 = gdp[(gdp['Duration'] =='2015-16') & (gdp['Items  Description']=='GSDP - CURRENT PRICES (` in Crore)')]
```

```python
gdp_2015_16 = gdp_2015_16.iloc[:,2:34].transpose()
```

```python
gdp_2015_16=gdp_2015_16.rename(columns={4:'Total GDp of states in 2015-16'})
```

```python
gdp_2015_16 = gdp_2015_16.dropna()
```

```python
gdp_2015_16
```

```python
plt.figure(figsize=(10,8), dpi = 600)

sns.barplot(x = gdp_2015_16['Total GDp of states in 2015-16'], y = gdp_2015_16.index,palette='plasma')
plt.xlabel('Total GDP of States for 2015-16', fontsize=12)
plt.ylabel('States', fontsize=12)
plt.title('Total GDP of States 2015-16 for all the states',fontsize=12)
plt.show()
```
